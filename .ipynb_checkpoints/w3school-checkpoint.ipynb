```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.grid()
plt.title("Sport watch data",fontdict=font1,loc='left')
plt.xlabel("Average pulse",fontdict=font2,loc='left')
plt.ylabel("Calorie burnage",fontdict=font2,loc='bottom')
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 70, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,linewidth=2,ls='--',marker='o',ms=10,mec='c',mfc='y')
plt.grid(color='r',linewidth=1)
plt.title("Sport watch data",fontdict=font1,loc='left')
plt.xlabel("Average pulse",fontdict=font2,loc='left')
plt.ylabel("Calorie burnage",fontdict=font2,loc='bottom')
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2,2,1)
plt.title("Graph1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])
plt.subplot(2,2,2)
plt.title("Graph2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(a,b)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2,2,3)
plt.title("Graph3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])
plt.subplot(2,2,4)
plt.title("Graph4")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(a,b)

plt.subplots_adjust(left=0.1,
                    bottom=1, 
                    right=0.9, 
                    top=2, 
                    wspace=0.4, 
                    hspace=0.4)

plt.suptitle("Shop")
plt.show()
```
